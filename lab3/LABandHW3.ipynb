```python
import numpy as np
from skimage import data
import matplotlib.pyplot as plt
import random
```

```python
image = data.camera()
for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        image[i,j] += random.randint(-20,20)
```

```python
plt.imshow(image, cmap = plt.cm.gray)
```

```python
mm = np.median(image)
print(mm)
```

```python
binary = image > mm
```

```python
plt.imshow(binary, cmap = plt.cm.gray)
```

```python
from skimage import feature

edges1 = feature.canny(image,sigma=2)
```

```python
plt.imshow(edges1, cmap = plt.cm.gray)
```

```python
def auto_canny(image, sigma=2):
	# compute the median of the single channel pixel intensities
	v = np.median(image)
	# apply automatic Canny edge detection using the computed median
	lower = int(max(0, (1.0 - sigma) * v))
	upper = int(min(255, (1.0 + sigma) * v))
	edged = feature.canny(image, lower, upper)
	# return the edged image
	return edged
```

```python
auto = auto_canny(image)
plt.imshow(auto, cmap = plt.cm.gray)
```
